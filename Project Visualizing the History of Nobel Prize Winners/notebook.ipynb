The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Start coding here!
data = pd.read_csv("data/nobel.csv")
pd.set_option('display.max_columns', None)
print(data.head())
```

### Q1: What is the most commonly awarded gender and birth country?⬇️

```python
top_gender = data['sex'].mode()[0]
top_country = data['birth_country'].mode()[0]
print(top_gender)
print(top_country)
```

### Q2: Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?⬇️

 creating a decade column ⬇️

```python
data['decade'] = (data["year"] // 10) * 10
```

Count total winners per decade ⬇️

```python
total_per_decade = data.groupby("decade").size()
```

Count US-born winners per decade ⬇️

```python
us_per_decade = data[data['birth_country']== 'United States of America'].groupby('decade').size()
```

Compute the ratio ⬇️

```python
ratio = us_per_decade / total_per_decade
```

```python
max_decade_usa = ratio.idxmax()
print(max_decade_usa)
```

### Q3: Which decade and Nobel Prize category combination had the highest proportion of female laureates?⬇️

Subset only female winners ⬇️

```python
data['female_winners'] = data['sex'] == 'Female'
```

```python
# Ensure 'decade' and 'category' columns exist in the 'data' DataFrame
if 'decade' in data.columns and 'category' in data.columns:
    # Subset only female winners
    female_winners = data[data['female_winners']]
    
    # Group by 'decade' and 'category' and calculate the proportion of female winners
    female_ratio = female_winners.groupby(['decade', 'category']).size() / data.groupby(['decade', 'category']).size()
    
    # Find the decade and category combination with the highest proportion of female laureates
    max_female_ratio = female_ratio.idxmax()
    
    # Store the result as a dictionary
    max_female_dict = {max_female_ratio[0]: max_female_ratio[1]}
    print(max_female_dict)
else:
    raise KeyError("Columns 'decade' and/or 'category' not found in the main DataFrame")
```

# Q3:Who was the first woman to receive a Nobel Prize, and in what category?

```python
female_winners = data[data['sex']=='Female']
first_woman = female_winners.sort_values("year", ascending=True).iloc[0]
first_woman_name = first_woman['full_name']
first_woman_category = first_woman['category']
print(first_woman_name)
print(first_woman_category)
```

## Q4/Which individuals or organizations have won more than one Nobel Prize throughout the years?


```python
# Individuals who won more than once
individuals = data[data["laureate_type"] == "Individual"]
repeat_individuals = individuals["full_name"].value_counts()
repeat_individuals = repeat_individuals[repeat_individuals >= 2]

# Organizations who won more than once
organizations = data[data["laureate_type"] == "Organization"]
org_names = organizations["organization_name"].fillna(organizations["full_name"])
repeat_organizations = org_names.value_counts()
repeat_organizations = repeat_organizations[repeat_organizations >= 2]


# Combine both as a single list
repeat_list = list(repeat_individuals.index) + list(repeat_organizations.index)
print(repeat_list)
```
